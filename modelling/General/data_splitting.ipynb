```python
from shutil import copy2
from sklearn.model_selection import train_test_split
import os
import numpy as np
import cv2
from tqdm import tqdm
```

```python
data_path=       "/media/kashraf/Elements/Dissertation/data/preprocessed/audio/stack/interv7/full"
dest_train_path= "/media/kashraf/Elements/Dissertation/data/preprocessed/audio/stack/interv7/train"
dest_test_path= "/media/kashraf/Elements/Dissertation/data/preprocessed/audio/stack/interv7/test"
```

```python
def split(path,train_path,test_path,cl_level):
    fnames=os.listdir(os.path.join(path,cl_level))
    
    f_train,f_test,yt,yes=train_test_split(fnames,range(len(fnames)),test_size=0.3)
    
    for file in tqdm(f_train):
        copy2(os.path.join(path,cl_level,file),train_path+"/"+cl_level)
    for file in tqdm(f_test):
        copy2(os.path.join(path,cl_level,file),test_path+"/"+cl_level+"/")
```

```python
split(data_path,dest_train_path,dest_test_path,"cl2")
split(data_path,dest_train_path,dest_test_path,"cl4")
split(data_path,dest_train_path,dest_test_path,"cl6")
split(data_path,dest_train_path,dest_test_path,"cl8")
```

```python

def gen_interv_data(cl_level):
    for i in tqdm(range(1,8)):
        path = os.path.join(data_path,"interv_"+str(i),"full")
        
#     print(os.path.join(path,cl_level))
        train_path = os.path.join(data_path,"interv_"+str(i),"train")
        print(train_path)
        test_path  = os.path.join(data_path,"interv_"+str(i),"test")
        split(path,train_path,test_path,cl_level)
        
gen_interv_data("cl2")
gen_interv_data("cl4")
gen_interv_data("cl6")
gen_interv_data("cl8")


    
```
